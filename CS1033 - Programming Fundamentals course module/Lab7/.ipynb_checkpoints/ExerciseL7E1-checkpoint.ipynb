```python
fo=open(input()) 

details=[]

for line in fo:
    Line = [i for i in line.strip().split(" ")] #put details to a list line by line
    while '' in Line:
        Line.remove('') #remove unnecessary spaces in each line
    details.append(Line)

name = []
birthday = []
gender = []

for i in details:
    #separate details to lists in order
    name.append(i[0])
    birthday.append(i[1])
    gender.append(i[2][:1])

def days_to_birthday(date):
    days_for_month=[31,28,31,30,31,30,31,31,30,31,30,31] #no of days in each month
    
    if(int(date[:4])%4 == 0): #leap year
        days_for_month[1] = 29 #change days in February
    
    no_of_months = int(date[5:7])-1  #count no of months to birthday
    
    Sum = 0
    for i in range(1, no_of_months+1):
        Sum = Sum + days_for_month[i-1] 
    
    #get the index of date to check the gender
    #use iterator to find the corresponding index of equivalent items
    date_index = birthday.index(date,iterator) 
    
    if(gender[date_index] == 'F'): #If female add 500
        totalDays = Sum + int(date[8:]) + 500
    else:
        totalDays = Sum + int(date[8:])
       
    
    if(totalDays<10):
        totalDays = '00'+str(totalDays)
    elif(10<=totalDays<100):
        totalDays = '0'+str(totalDays)
    else:
        totalDays = str(totalDays)
    
    return totalDays

#find order of submission
Year = [] #list with years of each birthday

for date in birthday:
    Year.append(date[:4])

Set_of_Year = set(Year) #set containing unique years
counter = []
for u_year in Set_of_Year:
    a = 1
    counter1 = [] #list with counts of each year
    for y in Year:
        #count one year at one time
        if(u_year == y):
            counter1.append(a)
            a = a + 1
        else:
            counter1.append(0)
    counter.append(counter1) #put list of counts of each year to a list 
Total = []
for i in range(len(Year)):
    t=0
    for j in range(len(Set_of_Year)):
        t = t + counter[j][i] #add elements column by column,because we should represent all counts in one list
    Total.append(t)
    
orderOfSub = []
for number in Total:
    if(number<10):
        number = "00"+str(number)
        orderOfSub.append(number)
    elif(10<=number<100):
        number = "0"+str(number)
        orderOfSub.append(number)
    else:
        orderOfSub.append(str(number))

#find NIC number
ID=[]
global iterator
iterator=0
for date in birthday:
    ID.append(date[:4] + days_to_birthday(date) + orderOfSub[birthday.index(date,iterator)])
    iterator=iterator+1
        

outFile = open("output.txt","a")
for j in range(len(details)):
    s = name[j]+" "+ID[j]
    outFile.write(s)
    if(j != len(details)-1):
        outFile.write("\n")

outFile.close()       
```
