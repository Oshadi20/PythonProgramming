```python
dimension = [int(i) for i in input("Enter the dimension: ").split(",")]
no_rows = dimension[0]
no_columns = dimension[1]
global A,B
A = []
B = []
def read(s):
    print("Enter Matrix "+ s +':')
    for i in range(no_rows):
        rows=[]
        row= input()
        for j in row.split():
            rows.append(int(j))
            
        if(s=='A'):
            A.append(rows)
        else:
            B.append(rows)
        

def transpose(m):
    global T
    T=[]
    for i in range(no_columns):
        I=[]
        for j in range(no_rows):
            I.append(m[j][i])
        T.append(I)
    return T


def product(m1,m2):
    global matProduct
    matProduct=[]
    for i in range(no_rows):
        mat=[]
        for k in range(no_rows):
            add=0
            for j in range(no_columns):
                add=add+m1[i][j]*m2[j][k]
            mat.append(add)    
        matProduct.append(mat) 
    
def display(m3):
    productMatrix=''
    for i in m3:
        r=''
        for j in i:
              r=r+str(j)+' '
        productMatrix=productMatrix+r+'\n'
    print(productMatrix)


try:
    read('A')
    AT = transpose(A)
except ValueError:
    print("Error")
except IndexError:
    print("Invalid Matrix")
else:
    read('B')
    try:
        BT=transpose(B)
        product(A,BT)
    except ValueError:
        print("Error")
    except IndexError:
        print("Invalid Matrix")
    else:
        display(matProduct)
```
